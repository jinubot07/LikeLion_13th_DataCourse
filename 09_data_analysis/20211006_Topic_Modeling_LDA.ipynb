```python
from gensim import corpora
from gensim import models
```

```python
documents=[
  '    .    .',
  '      .     .',
  '    .',
  '       . ',
  '         .'
]
```

```python
stoplist = ('.','!', '?','','')  #   
doc1 = '     .'

# print(doc1.split())
tex_list = []
for word in doc1.split():
  if word not in stoplist:
    tex_list.append(word)
tex_list
```

```python
stoplist = ('.','!', '?', '', '')  #   

#    
texts = [
         [word for word in document.split() if word not in stoplist]  for document in documents
        ]
texts
```

```python
dictionary = corpora.Dictionary(texts)
print(type(dictionary) , list(dictionary) )
```

```python
#  ()
# doc2bow()     .
#  ID    .

corpus = [dictionary.doc2bow(text) for text in texts]
print(type(corpus), 'corpus : {}'.format(corpus))
```

```python
lda = models.ldamodel.LdaModel( corpus=corpus,
                                id2word=dictionary,
                                num_topics=2,
                                random_state = 1)

print(type(lda))
```

```python
# def show_topics(num_topics=10, num_words=10, log=False, formatted=True)
lda.show_topics()
```

```python
print( lda.num_terms )     ##  
print( lda.get_topics() )  ## topic  
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
path = '/usr/share/fonts/truetype/nanum/NanumGothicEco.ttf'
```

```python
wc = WordCloud(background_color='white',
               max_words=200,
               contour_width=3,
               contour_color='steelblue', 
               font_path=path)
```

```python
fig = plt.figure(figsize=(30,30))

for t in range(lda.num_topics):
  plt.subplot(5,4,t+1)
  x = dict(lda.show_topic(t,200))    #  200  
  im = wc.generate_from_frequencies(x)
  plt.imshow(im)
  plt.axis("off")
  plt.title("Topic #" + str(t))

plt.savefig('LDA_wordcloud.png', dpi=300, bbox_inches='tight')
plt.show()
plt.close(fig)
```
