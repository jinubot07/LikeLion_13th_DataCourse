```python
# 꼬꼬마 - 명사확인
# from konlpy.tag import Kkma 
# k = Kkma()

# Okt - 명사확인
from konlpy.tag import Okt   ### Okt
o = Okt()

doc_nouns = o.nouns(doc_mam)
print( len( doc_nouns ) )
doc_nouns[0:500]
```

```python
import nltk
```

```python
# nltk.Text()를 이용하여 nltk가 가지는 많은 기능을 사용 가능해짐.
ko = nltk.Text(doc_nouns, name="맘마미아1")
print(type(ko))
print(len(ko.tokens))
```

```python
most_fre = ko.vocab().most_common(50)
most_fre
```

```python
# 중복된 단어를 제거한 개수를 확인
print(len(set(ko.tokens)))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12, 6))
ko.plot(50)
plt.show()
```

```python
%matplotlib inline
import matplotlib as mpl # 기본 설정 만지는 용도
import matplotlib.pyplot as plt # 그래프 그리는 용도
import matplotlib.font_manager as fm # 폰트 관련 용도
import numpy as np

path = '/usr/share/fonts/truetype/nanum/NanumGothicEco.ttf' # 설치된 나눔글꼴중 원하는 녀석의 전체
font_name = fm.FontProperties(fname=path, size=10).get_name()
print(font_name)
plt.rc('font', family=font_name)

## 음수 표시되도록 설정
mpl.rcParams['axes.unicode_minus'] = False 
```

```python
# 한글에서는 따로 불용어 사전이 없어, 따로 만들거나 또는 파일로 부터 불러올 수 있다.
stop_words = [ '내', '년','점','분','편','이','그', '것', '또', '좀', '수', '좀', '더', '낼', '거','도', '부분', '나',
              '다음', '그냥', '말', '뭐', '애', '나', '듯', '편', '볼', '점', '중', '로']

new_ko = [ ]
for one_word in ko:
  if one_word not in stop_words:
    new_ko.append(one_word)
```

```python
### nltk Text 객체 만들기
new_ko = nltk.Text(new_ko, name="맘마미아1")
plt.figure(figsize=(12,6))
new_ko.plot(50)
```

```python
plt.figure(figsize=(15,8))
new_ko.dispersion_plot(['스토리', '대박', '뮤지컬'])
```

```python
from wordcloud import WordCloud, STOPWORDS

import numpy as np
from PIL import Image
```

```python
mammamia_mask = np.array(Image.open("mammamia.png"))
```

```python
data = new_ko.vocab().most_common(1000)
```

```python
### 워드 클라우드 표현을 위한 데이터 생성
### 약간의 시간이 필요.
wc = WordCloud(background_color='white', 
               max_words=200,
               mask=mammamia_mask,   
               contour_width=3, 
               contour_color="steelblue", 
               font_path=path).generate_from_frequencies(dict(data))
```

```python
plt.figure(figsize=(12,8))
plt.imshow(wc)
plt.axis("off")
plt.show()
```
