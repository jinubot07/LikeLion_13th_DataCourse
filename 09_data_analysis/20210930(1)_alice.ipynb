### 꼬꼬마를 이용한 분석
  * 문장 분석
  * 명사 분석
  * 형태소 분석

#### 문장 분석

```python
# 문장 단위로 리스트에 배열
k = Kkma()
k.sentences("안녕하세요! 오늘은 한글 분석을 시작합니다.")
```

#### 명사 분석

```python
# 명사 단위로 리스트에 배열
k.nouns("안녕하세요! 오늘은 한글 분석을 시작합니다.")
```


#### 형태소 분석
[http://kkma.snu.ac.kr/documents/index.jsp?doc=postag](http://kkma.snu.ac.kr/documents/index.jsp?doc=postag)



```python
# 형태소 단위로 리스트에 배열
k.pos("안녕하세요! 오늘은 한글 분석을 시작합니다.")
```

```python
import numpy as np
from PIL import Image
```

```python
img = Image.open('alice_color.png') #  불러온 이미지를 변수명 img에 저장하기
type(img), img
```

```python
alice_mask = np.array(img)
alice_mask
```

```python
alice_mask.shape
```

```python
plt.figure(figsize=(15,8))
plt.imshow(alice_mask)
plt.imshow(alice_mask, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
hany = Image.open('하니.png') #  불러온 이미지를 변수명 img에 저장하기
plt.figure(figsize=(15,8))
plt.imshow(hany)
plt.axis('off')

plt.show()
```

```python
# 객체 생성
wc = WordCloud(max_words = 2000,
               mask = alice_mask,
               background_color ='white', # 배경색
               contour_width = 3, # 외곽선두께
               contour_color = 'steelblue') # 외곽선 색
# 생성된 객체를 이용해서 데이터 넣기
wc.generate(text)
```

```python
plt.figure(figsize=(15,8))
# 이미지 손실이 복구하는 법에 대한 정의
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')

plt.show()
```
