```python
tips = sns.load_dataset("tips")
type(tips)
```

```python
tips(head)
```

### 1) 요일별 식사금액은 얼마나 될까?

```python
sns.boxplot(x='day', y='total_bill', data=tips)
plt.show()
```

#### 1) 알아낸 정보
* 식사금액은 토요일과 일요일이 많구나
* 현재 데이터는 목,금,토,일만 있구나
* 토요일과 일요일을 비교해 본다면 상대적으로 일요일 지출 식사금액이 많다.

### 2) 요일별 팁에 대해서 알아보자.

```python
tips.columns
```

```python
sns.boxplot(x='day', y='tip', data=tips)
```

#### 2) 알아낸 정보
* 일요일에 팁이 가장 많다.
* 금요일의 중앙 값이 높은 편이다.  -> 왜 높을까?

### 3) 요일별 사이즈(식사인원)에 대해 알아보자

```python
sns.boxplot(x='day', y='size', data=tips)
```

#### 3) 알아낸 정보
* 목요일에 이상값이 가장 많다.
* 토요일과 일요일에는 식사 인원이 많다.  -> (첫번째 그래프 해석) 식사인원이 많기 때문에 식사 금액이 상대적으로 높을 수 밖에 없다.

주의! x축 : 범주형 (수치형X)

### 4) 누가 팁을 더 많이 주는가?

```python
sns.boxplot(x='day',y='tip',hue='sex',data=tips)
```

#### 4) 알아낸 정보
* 남성의 팁은 편차가 크다. 많이 주는 사람도 있고, 적게 주는 사람도 있다.
* 토요일은 이상치가 가장 많다.

### 5) 흡연과 팁 의 관계

```python
sns.boxplot(x='day',y='tip', hue='smoker', data=tips)
```

#### 5) 알아낸 정보
* 흡연 여부와 팁과는 특이점이 없다.

### 6) 시간과 팁의 관계

```python
sns.boxplot(x='day',y='tip', hue='time', data=tips)
```

```python
sns.boxplot(x='time',y='tip', hue='day', data=tips)
```

### 7) 식사금액과 팁의 상관관계
#### Regression(회귀선)을 그어서 대략적인 예측을 수행해보자

```python
sns.set_style('darkgrid')
sns.lmplot(x='total_bill', y='tip', data=tips, size=7)
plt.show()
```

```python
sns.set_style('darkgrid')
sns.lmplot(x='total_bill', y='tip', hue='smoker', data=tips, size=7)
plt.show()
```

### 8) 연도별 월별 비행기 승객 인원 수

```python
fg = sns.load_dataset('flights')
fg
```

```python
sns.scatterplot(x='year', y='passengers',data=fg)
```
