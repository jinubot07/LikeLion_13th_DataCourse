```python
# 방법 1
import time
time_start = time.time()

for i in range(99):
  folium.Marker(location = [dat_tmp.loc[i,'위도'],dat_tmp.loc[i,'경도']],
                popup   = dat_tmp.loc[i,'name'],
                icon    = folium.Icon(color='red', icon='info-sign')
                ).add_to(m)
time_work = time.time() - time_start

print(time_work)
m
```

```python
# 방법 2

des =dat_tmp['name']
dat_p1 = list(dat_tmp['위도'])
dat_p2 = list(dat_tmp['경도'])
des

dat_xy=list(zip(dat_p1,dat_p2))
dat_xy[0:3]
```

```python
time_start = time.time()

m = folium.Map(location = [37.492386, 126.909832], zoom_start= 13)

for name, loc in zip(des, dat_xy):
  folium.Marker(location=loc,
                tooltip = name,
                icon = folium.Icon(color='blue', icon='info-sign')
                ).add_to(m)

m.save("seoul_toilet_100.html")

time_work = time.time() - time_start
print(time_work)
m
```

```python
# 일부 데이터만 가져오기
dat_park = dat_bike.loc[4:13, ['보관소(대여소)명','Unnamed: 4','Unnamed: 5']]
# 컬러명 바꾸기
dat_park.columns = ['지역명','위도','경도']
dat_park
```
