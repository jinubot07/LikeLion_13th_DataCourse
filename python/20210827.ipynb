```python
age = int(input("나이 입력 : "))
print("당신의 나이는 ", age, "입니다.", sep="\"")
# sep함수가 언급X -> 기본값 스페이스 한 칸
# 큰 따옴표를 문자열안에 넣기 위해선 직전에 역슬래쉬를 넣어준다.
```

```python
# end함수 : print()호출시 마지막 문자로 쓰이는 문자, 기본값으로 \n(개행)이 설정 되어있다.

print(1,end = " ")
print(3,end = " ")
print(2,end = " ")
```

```python
# 산술 연산자
+, -, *, /, *, //(몫), %(모듈러,나머지), **(거듭제곱)
# 비교 연산자
== 같다(결과값 : True,False), != 다르다, >= 크거나 같다, <= 작거나 같다
자료형 : bool(불리언Boolean)
# 논리 연산자
and (이항 연산자), or (이항 연산자), not (단항 연산자)
A and B : A 또는 B가 False면 False
A or B : A 또는 B가 True면 True
not A : A 가 True면 False

```

```python
gender = input("성별 입력(남/녀) : ")
age = input("나이 입력(숫자) : ")
today = input("오늘 날짜(15/30) : ")

if(gender == "녀" and int(age) >= 18 and today == "15") :
  print("환영합니다.")
elif (gender == "남" and int(age) >= 18 and today == "30") : 
     print("환영합니다.")
else :
        print("조건을 다시한번 확인해주세요.")
```
