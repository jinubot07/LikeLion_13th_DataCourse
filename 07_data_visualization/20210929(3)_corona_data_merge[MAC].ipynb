# 생성한 데이터 셋을 불러와 작업하기

```python
import pandas as pd

# excel 데이터 셋도 pd.read_excel로 가져올 수 있음.
corona = pd.read_csv("./data_set/2021-09-28_corona.csv")
vaccine = pd.read_csv("./data_set/20210929_14_vaccine_bloomberg.csv")
country_code = pd.read_csv("./data_set/country.csv",encoding='cp949')

corona.shape, vaccine.shape, country_code.shape
```

### corona 데이터와 country_code 데이터 합치기

```python
corona.head()
```

```python
corona['한글표기']= corona['국가'].str.extract('([ㄱ-ㅣ가-힣]+)',expand=False)
corona.columns
```

```python
col = ['국가', '한글표기', '위중증합계', '위중증1일', '치명(%)', '완치(%)', '발생률', 
       '인구수', '확진자합계', '확진자1일', '사망자합계', '사망자1일', '완치합계', '완치1일']

new_corona = corona[col].copy()
new_corona
```

```python
country_code.head()
```

```python
country_code.columns = ['han_code','eng_code','country','etc']
country_code.head()
```

```python
# 데이터 셋끼리 합치기
df_corona = new_corona.merge(country_code, left_on = '한글표기', right_on = 'han_code')
df_corona.head()
```

```python
# 데이터 셋에 불필요한 컬럼명 삭제
df_corona.drop(['국가','han_code','country','etc'],axis=1, inplace=True)
df_corona.head(10)
```

### corona + country_code 데이터와 vaccine 데이터 합치기

```python
df_corona_all = df_corona.merge(vaccine, left_on='eng_code', right_on='국가')
df_corona_all.head()
```

```python
# 한글표기 와 국가 데이터 내용이 겹친다. 국가 컬럼을 없애버리자
df_corona_all.drop(['국가'],axis=1,inplace=True)
df_corona_all
```
