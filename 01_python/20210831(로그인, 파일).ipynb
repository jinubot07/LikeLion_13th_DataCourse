```python
# 소문자 및 특수문자를 추가 암호화
import random

spc1_str = ""  # 특수문자 문자열 
for x in range(33,48): # 33 : !, 47 : /
    spc1_str += chr(x)
print('        특수문자로 구성된 문자열1:', spc1_str)

spc2_str = ""  # 특수문자 문자열 
for x in range(58,65): # 58 : : , 64 : @
    spc2_str += chr(x)
print('        특수문자로 구성된 문자열2:', spc2_str)

small_str = "" # 소문자
for x in range(97,123): # 97 : a, 122 : z
    small_str += chr(x)
print('          소문자로 구성된 문자열 :', small_str)

pwd_str = spc1_str + spc2_str + small_str # 특수문자 ~ 소문자로 구성된 문자열
print('특수문자~소문자로 구성된 최종 문자열:', pwd_str)

def passGen():
  password = ""
  for i in range(20):
    idx = random.randrange(len(pwd_str))
    password = password + pwd_str[idx]  # 특수문자 ~ 소문자로 구성된 문자열에서 특정 문자로 구성하여 암호화
  return password

print('                   생성된 암호:', passGen())
```

```python
### 이미지 프로그래밍
# 원본파일 복사
file1 = "smalldog.jpeg" #원본
backup = "smalldog_back.jpeg" #사본

infile = open(file1,'rb')
outfile = open(backup, 'wb')

while True : # 파일읽는것을 마무리 할 때 까지
  copy_buffer = infile.read(1024)

  if not copy_buffer : # 읽을게 없다면
    break # 동작그만

  outfile.write(copy_buffer)

infile.close()
outfile.close()
print("복사완료")
```

```python
### 실습 
# 파일 하나의 복사본 2개씩 만든다
# 두개의 파일명을 지정. test.jpeg -> test01.jpeg, test02.jpeg

file1 = "smalldog.jpeg" #원본
backup1 = "smalldog01.jpeg" #사본
backup2 = "smalldog02.jpeg" #사본

# file1을 backup1로 복사
infile = open(file1,'rb')
outfile = open(backup1, 'wb')
while True : # 파일읽는것을 마무리 할 때 까지
  copy_buffer = infile.read(1024)
  if not copy_buffer : # 읽을게 없다면
    break # 동작그만
  outfile.write(copy_buffer)

# file1을 backup2로 복사
infile = open(file1,'rb')
outfile = open(backup2, 'wb')

while True : # 파일읽는것을 마무리 할 때 까지
  copy_buffer = infile.read(1024)
  if not copy_buffer : # 읽을게 없다면
    break # 동작그만
  outfile.write(copy_buffer)

infile.close()
outfile.close()
print("복사완료")
```

```python
# 파일명을 지정. test.jpeg -> test01.jpeg, test02.jpeg

origin = "smalldog.jpeg" #원본
cnt = input("사본 수:")

for i in range(int(cnt)) :
  filename = "smalldog_" + str(i+1)  + '.jpeg' #사본
  backup = filename

  # origin을 backup로 복사
  infile = open(origin,'rb')
  outfile = open(backup, 'wb')

  while True : # 파일읽는것을 마무리 할 때 까지
    copy_buffer = infile.read(1024)
    if not copy_buffer : # 읽을게 없다면
        break # 동작그만
    outfile.write(copy_buffer)



infile.close()
outfile.close()
print("복사완료")
```
