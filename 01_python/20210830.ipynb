```python
# 5-4 실습 - 원하는 음식을 q가 나올때까지 추가(append), 추가 후, 다시 빼고 싶은 음식을 빼준다

list=[]
while True :
  food = input("좋아하는 음식 : ")
  if food == "q":
    break
  # 같은 메뉴가 이미 있으면 예외처리
  elif food in list: 
    print("이미 리스트에 있는 메뉴 입니다.")
  # 입력값 크기제한
  elif len(food) >= 20 :
    print("글자의 길이가 너무 깁니다.")
  else :
    list.append(food)

while True :
  omit = input("빼고싶은 음식 : ")
  if omit == "q":
    break
  # 예외처리
  elif omit not in list :
    print("리스트에 없는 메뉴 입니다.")
  # 입력값 크기제한
  elif len(omit) >= 20 :
    print("글자의 길이가 너무 깁니다.")
  else :
    list.remove(omit)


print(list)
```
